```sage
# 1a
reset()

t=var('t')
k=var('k')
x0=var('x0')
x=function('x')(t)
eqd = diff(x,t)==-k*x
sol = desolve(eqd,[x,t],[0,x0])
sol
```

```sage
# 1b
plot([sol(t,0.1,10),sol(t,0.5,10),sol(t,1,10)],t,0,3)
```

```sage
# 1c
t12 = 5730
eq1 = sol(t12, k, x0) - x0/2 == 0
k_jum = solve(eq1, k)[0].rhs()
k_jum
```

```sage
# 1d
ec = sol(t,k_jum,1)==0.2
solve(ec,t)
```

```sage
# 1e

t1 = solve(sol(t,k_jum,1) ==0.9157, t)
t2 = solve(sol(t,k_jum,1) ==0.93021, t)
sow(t1, " - ", t2)
```
