```python
# 1a
reset()

t=var('t')
k=var('k')
x0=var('x0')
x=function('x')(t)
eqd = diff(x,t)==-k*x
sol(t,k,x0) = desolve(eqd,[x,t],[0,x0])
sol
```

```python
# 1b
plot([sol(t,0.1,10),sol(t,0.5,10),sol(t,1,10)],t,0,3)
```

```python
# 1c
t12=var('t12')
eq1 = sol(t12, k, x0) == x0/2
fk(t12) = solve(eq1,k)
k_jum=fk(5730)[0].rhs()
k_jum
```

```python
# 1d
ec = sol(t,k_jum,1)==0.2
solve(ec,t)
```
