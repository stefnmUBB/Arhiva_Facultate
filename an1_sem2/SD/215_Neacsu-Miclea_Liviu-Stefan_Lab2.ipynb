```python
# 2 a
reset()
x=var('x')
y=function('y')(x)
eqd = diff(y,x) == 1+y^2
sol = desolve(eqd,y,[0,1])
show(sol)
f(x)=solve(sol,y(x))[0].rhs()
#yy=var('yy')
plot(f(x),x,-1,1,detect_poles='True',ymin=-10,ymax=10)
#implicit_plot(,(x,-1,1),(y,-1,1))
```

```python
# 3 a
reset()
x=var('x')
k = var('k')
y=function('y')(x)
eqd = diff(y,x) + k/x*y(x) == x^3
sol = desolve(eqd,[y,x])
show(sol)
```

```python
#3 b
C = var("_C")
sol1 = sol.substitute(C==1,k==1)
sol2 = sol.substitute(C==2,k==1)
sol3 = sol.substitute(C==3,k==1)
plot([sol1, sol2, sol3], x, -1, 1,ymin=-10,ymax=10)
```

```python
#3 c
sol = desolve(eqd,[y,x],ics=[1,0])
plot(sol.substitute(k==1),x,-1, 1, ymin=-10, ymax=10)
```

```python
#3 d
sols = [plot(sol.substitute(k==i),x,-1,1,ymin=-5,ymax=5) for i in [-3..-1]]
sols += [plot(sol.substitute(k==i),x,-1,1,ymin=-5,ymax=5) for i in [1..5]]
animate(sols)
```

```python
# 4
reset()
x=var('x')
a,b = var('a,b')
y=function('y')(x)
eqd = diff(y,x) == y(x)+a*x+b
sol = desolve(eqd,[y,x],[0,1])
resab = solve([sol(a,b,1)==-5+4*e, sol(a,b,2)==-7+4*e^2],a,b)[0]
f(x) = sol.substitute(a==resab[0].rhs(), b==resab[1].rhs())
plot(f,x,-1,1)
```
