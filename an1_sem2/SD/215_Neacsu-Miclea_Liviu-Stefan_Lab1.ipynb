```python
# 8 a)
t,s = var('t,s')
f(t,s) = 1-(s*cos(4*t)*cos(t))/sqrt(1-s^2 * cos(4*t)^2 * sin(t)^2)
x(t,s) = f(t-pi/2,s)
y(t,s) = f(t,s)
parametric_plot((x(t,1/2),y(t,1/2)),(t,0,2*pi))
```

```python
# 8 b)
g = parametric_plot((x(t,1/10),y(t,1/10)),(t,0,2*pi))

for s in range(2,11):
    g += parametric_plot((x(t,s/10),y(t,s/10)),(t,0,2*pi))
    
show(g)
```
