```python
# 2a
reset()
k=var('k')
t=var('t')
Tm=var('Tm')
T0=var('T0')

T=function('T')(t)
eqd = diff(T,t)==-k*(T(t)-Tm)
sol(t,k,Tm,T0) = desolve(eqd,[T,t],[0,T0])
sol
```
