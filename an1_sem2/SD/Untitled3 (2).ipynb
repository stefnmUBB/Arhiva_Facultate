```python
# 3c

t=var('t')
r=var('r')
x0=var('x0')

x=function('x')(t)

eq_malthus = diff(x,t)==r*x(t)
sol_m(t,r,x0) = desolve(eq_malthus, [x,t], [0,x0])

sol_m
```

```python

r0 = var('r0')
K=var('K')

eq_verhulst = diff(x,t)==r0*x(t)*(1-x(t)/K)
sol_v(t,r0,K,x0) = desolve(eq_verhulst, [x,t], [0,x0])

ec1 = sol_v.log_simplify()

sol_v(t,r0,K,x0) = solve(ec1,x)[0].rhs()
sol_v
```

```python
show(plot([sol_m(t,1,10),sol_m(t,2,10)],t,0,3))
```
